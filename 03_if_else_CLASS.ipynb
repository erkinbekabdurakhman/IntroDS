## Условный оператор

Вспомним про условные выражения (сравнения).

Ставим выбор перед двумя альтернативами.

```python
"""
if Условие:
    Блок инструкций 1
else:
    Блок инструкций 2
"""
# Блок инструкций 1 будет выполнен, если Условие истинно.
# Если Условие ложно, будет выполнен Блок инструкций 2.
```

```python
a = -5
if a > 0:
    # отступ
    # в юпитере tab = 4 пробела
    print("Это число положительное")
else: # иначе
    print("Это число отрицательное")
```

В условной инструкции может отсутствовать слово `else` и последующий блок. Такая инструкция называется неполным ветвлением.

Для выделения блока инструкций, относящихся к инструкции `if` или `else` в языке Питон используются отступы. Все инструкции, которые относятся к одному блоку, должны иметь равную величину отступа, то есть одинаковое число пробелов в начале строки.

Отступ в 4 пробела или символ табуляции (Рекомендуется использовать отступ в 4 пробела).

### Напишите программу, которая принимает слово. 
1. Если длина слова больше 5-ти, выведите "Это слово имеет больше 5 символов".
2. Если количество символов равно 5, выведите "Пять!".
3. Иначе выведите "Это слово имеет меньше 5 символов".

```python
word = input()

if len(word) > 5:
    print('Это слово имеет больше 5 символов')
else:
    if len(word) == 5:
        print("FIVE!")
    else:
        print('Это слово имеет меньше 5 символов')
        
    print('Это конеец!')
```

### Пользователь вводит число (целое).

- Если равно 0, "Это ноль!"

В конце программы напечатать "Конец программы"

```python
num = int(input())

if num > 0:
    print("Это число положительное!")
elif num < 0:
    print("Это число отрицательное!")
else:
    print("NULL!")
```

```python
# так мы импортируем необходимые модули

import random
```

```python
# функция генерирует случайное число

```

```python
num = random.randrange(1, 100)

while True:
    # здесь напишите решение
    otvet = int(input())
    
    if otvet == num:
        print('Вы угадали!')
        break
    elif otvet > num:
        print('Ваше число больше загаданного')
    elif otvet < num:
        print('Ваше число меньше загаданного')
    
```
