## Несколько графиков

### В одном графике

```python
x = np.linspace(-10, 10, 200)
y1 = x**2
y2 = -x**2 + 20

plt.title("Зависимости: y1 = x^2, y2 = -x^2 + 20") # заголовок
plt.xlabel("x")         # ось абсцисс
plt.ylabel("y1, y2")    # ось ординат
plt.grid()              # включение отображение сетки
plt.plot(x, y1, x, y2)  # построение графика
```

### В разных графиках

```python
x = np.linspace(-10, 10, 200)
y1 = x**2
y2 = -x**2 + 20
```

```python
plt.figure(figsize=(9, 9))
# в качестве аргумента, мы передаем кортеж, определяющий размер общего поля. 


plt.subplot(2, 1, 1)

# первый аргумент – количество строк, 
# второй – столбцов в формируемом поле, 
# третий – индекс (номер поля, считаем сверху вниз, слева направо).


plt.plot(x, y1)               # построение графика
plt.title("Зависимости: y1 = x^2, y2 = -x^2 + 20") # заголовок
plt.ylabel("y1", fontsize=14) # ось ординат
plt.grid(True)                # включение отображение сетки

plt.subplot(2, 1, 2)
plt.plot(x, y2, "r-.")               # построение графика
plt.xlabel("x", fontsize=14)  # ось абсцисс
plt.ylabel("y2", fontsize=14) # ось ординат
plt.grid(True)                # включение отображение сетки
```
